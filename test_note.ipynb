```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_node(1) # 노드 1 추가
G.add_nodes_from([2, 3, 4]) # 여러 개 한 번에
pos = nx.spring_layout(G, k=0.1)
nx.draw(G, pos,with_labels=True, node_color="skyblue", node_size=40000, font_size=12)
plt.show()
```
